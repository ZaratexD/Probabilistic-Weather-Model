```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
A = np.array([[0.5, 0.5, 0.24],
              [0.25, 0.0, 0.25],
              [0.25, 0.5, 0.5]])
```

```python
print(A)
```

```python
x_today = np.array([[1], 
                    [0],
                    [0]])
```

```python
for k in range(50):
    x_tomorrow = A@x_today
    print(k)
    print(x_tomorrow)
    x_today = x_tomorrow
```
